# Imports and Basic Process

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./Weeki_Wachee_spring_10079u.jpg')
plt.imshow(image[...,::-1])
```

```python
"""
alpha_beta_process function:
    
    this funciton will create transformed output by given alpa beta coefficients and linear transforms for changing
    the contrast and brightness. 
    args -> 
        alpha: alpha coefficient will be used for contrast  tranforms
        gamma: beta coefficient will be used for brightness  tranforms
"""
def alpha_beta_process(image ,alpha , beta):
    new_image = cv.convertScaleAbs(image, alpha=alpha, beta=beta)

            
    plt.figure(figsize=[10,5])
    plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original image");
    plt.subplot(122);plt.imshow(new_image[...,::-1]);plt.title("Output image");
```

# Outputs

```python
alpha_beta_process(image ,0.2 , 40)
```

```python
alpha_beta_process(image ,2 , 40)
```

```python
alpha_beta_process(image ,2 , -100)
```

```python
alpha_beta_process(image ,3 , 0)
```
