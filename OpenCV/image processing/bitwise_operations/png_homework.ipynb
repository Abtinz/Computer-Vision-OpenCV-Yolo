# Imports

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image = cv.imread("./sherlock.jpg")
plt.imshow(image[:,:,::-1])
```

```python
#here we are building fundamental png image
png_result = np.zeros((1445,1445,4))
png_result[... , 0:3] = image
```

# alpha chennel filters for faces of Dr.Watson and Sherlock

```python
alpha_channel = np.full((1445,1445), 10)
cv.circle(alpha_channel, (580,150), 150, 255, -1) 
cv.circle(alpha_channel, (900,300), 150, 255, -1) 
plt.imshow(alpha_channel,cmap='gray');plt.title("alpha channel");
```

# Building the fourth channel of image for png format

```python
#here we will save sherlock and wotsons face with transparency channel in png formt
#we will use multi-level transparency filters for beter observation of this process
for i in range(10):
    alpha_channel = np.full((1445,1445), 10*i)
    cv.circle(alpha_channel, (580,150), 150, 255, -1) 
    cv.circle(alpha_channel, (900,300), 150, 255, -1) 
    png_result[... , -1] = alpha_channel
    cv.imwrite(f"result_{i}.png", png_result)
```
