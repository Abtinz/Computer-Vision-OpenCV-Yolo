```python
import cv2
import matplotlib.pyplot as plt
```

video reading from specified url

```python
video_source = cv2.VideoCapture("./test.mp4")

while True:
    ret, frame = video_source.read()
    
    if not ret:
        print("Exiting ...")
        break
    cv2.imshow('video', frame)
    if cv2.waitKey(1) & 0xFF == 27:
        plt.imshow(frame[...,::-1])
        break

video_source.release()
cv2.destroyAllWindows()
```
