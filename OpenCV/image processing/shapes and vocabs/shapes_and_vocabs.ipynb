```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lets create our simple black board which we are going to draw others on
black_board = np.zeros((500,500,3), "uint8")

plt.imshow(black_board[...,::-1]); plt.title("black_board")
```

# Shapes

```python
#cv.line(image or board, starting cordinates of line, ending cordinates of line, color, thickness)
cv.line(black_board, (0,0), (500,50), (0,255,255), 5)
cv.line(black_board, (0,0), (50,500), (140,200,255), 5)
cv.line(black_board, (0,0), (500,500), (255,255,100), 5)
plt.imshow(black_board[...,::-1]);plt.title("lines")
```

Notes:
    
    1. color parameters are in BGR mode
    
    2. coordintes are x,y not height and width

```python
#cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

cv.rectangle(black_board, (150,150), (400,250), (50,255,50), 8)

plt.imshow(black_board[...,::-1]); plt.title("Rectangle");
```

```python
#cv2.cirlce(image, center, radius, color, thickness)

cv.circle(black_board, (400, 400), 100, (30,20,200), 5) 
cv.circle(black_board, (400, 300), 100, (250,20,200), 5) 
cv.circle(black_board, (400, 200), 100, (200,120,100), 5) 
plt.imshow(black_board[...,::-1]); plt.title("Circle");
```

```python
cv.ellipse(black_board, (150, 400), (100, 50), 300, 0, 360, (100,100,100), 4)
cv.ellipse(black_board, (150, 400), (100, 50), 60, 0, 360, (100,200,100), 4)
cv.ellipse(black_board, (150, 400), (100, 50), 180, 0, 360, (0,255,0), 4)

plt.imshow(black_board[...,::-1]);plt.title("Ellipse");
```

```python
pts = np.array( [[10,10], [20,150], [100,100], [200,60]], np.int32)

pts = pts.reshape((-1,1,2))

cv.polylines(black_board, [pts], True, (0,0,255), 3)
pts = np.array( [[10,10], [20,70], [100,150], [200,5]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(black_board, [pts], True, (0,0,255), 3)

plt.imshow(black_board[...,::-1]);plt.title("Polygon");
```
