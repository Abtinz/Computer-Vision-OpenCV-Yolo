```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
black_board = np.zeros((500,500,3), "uint8")

plt.imshow(black_board[...,::-1]); plt.title("black_board")
```

```python
cv.rectangle(black_board, (100,200), (400,400), (50,255,50), -1)
cv.rectangle(black_board, (300,300), (350,390), (100,150,255), -1)
cv.rectangle(black_board, (120,220), (250,350),(255,100,100) , -1)

plt.imshow(black_board[...,::-1]); plt.title("Rectangle");
```

```python
pts = np.array( [[50,200], [250,50], [450,200]], np.int32)

pts = pts.reshape((-1,1,2))
cv.fillPoly(black_board, [pts], (0,0,255))

pts = np.array( [[120,150], [150,125], [120,100], [150,100]], np.int32)

cv.fillPoly(black_board, [pts], (100,150,255))

plt.imshow(black_board[...,::-1]);plt.title("Polygon");
```
