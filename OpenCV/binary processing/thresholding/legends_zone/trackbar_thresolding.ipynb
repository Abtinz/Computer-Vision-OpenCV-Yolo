```python
import cv2
import matplotlib.pyplot as plt
```

```python
trackbar_type = 'please select your Threshold type:\n0: Binary Threshold \n 1: Binary Threshold Inverted \n 2: Truncate \n 3: To Zero Threshold \n 4:Inverted To Zero  Threshold'
trackbar_value = 'trackbar_value'
window_name = 'Plate Binary Thresholding'

image = cv2.imread("plate.jpg" , 0)

def threshold_trackbar(val):
    global threshold_image, threshold_value, threshold_type
    
    threshold_type = cv2.getTrackbarPos(trackbar_type, window_name)
    threshold_value = cv2.getTrackbarPos(trackbar_value, window_name)
    _ , threshold_image = cv2.threshold(image, threshold_value, 255, threshold_type )

    cv2.imshow(window_name, threshold_image)
```

```python
cv2.namedWindow(window_name,cv2.WINDOW_NORMAL)
cv2.resizeWindow(window_name, 700, 600)
cv2.createTrackbar(trackbar_type, window_name , 0, 4, threshold_trackbar)
cv2.createTrackbar(trackbar_value, window_name , 0, 255, threshold_trackbar)


threshold_trackbar(0)
cv2.waitKey()
cv2.destroyAllWindows()

thresh_type= ["Binary Threshold","Binary Threshold Inverted ","Truncated","Threshold to Zero","Inverted To Zero  Threshold"]
plt.imshow(threshold_image, cmap='gray');plt.title("threshold valus: {} threshold method: {}".format(threshold_value, thresh_type[threshold_type]))
plt.show()
```
