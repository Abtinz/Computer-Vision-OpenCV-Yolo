```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./taken.jpg')
plt.imshow(image[...,::-1])
```

```python
alpha = float(input('alpha(0.1:3.0): '))
beta = int(input('beta(0:100): '))
```

output = image * alpha  + beta

alpha => contrast

bete => brightness

```python
new_image = cv.convertScaleAbs(image, alpha=alpha, beta=beta)

            
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original image");
plt.subplot(122);plt.imshow(new_image[...,::-1]);plt.title("Output");
```

```python
new_image2 = cv.convertScaleAbs(image, alpha=alpha, beta=beta)

            
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original image");
plt.subplot(122);plt.imshow(new_image2[...,::-1]);plt.title("Output-V2");
```

```python
new_image3 = cv.convertScaleAbs(image, alpha=alpha, beta=beta)

            
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original image");
plt.subplot(122);plt.imshow(new_image3[...,::-1]);plt.title("Output-V3");
```
