```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
color = (255,255,255)

def triangle(event,x,y,flags,param):
    global color
    if event == cv2.EVENT_LBUTTONDOWN:
        
        vertices = np.array([[x, y - 20], [x - 20 , y + 20], [x + 20, y + 20]], np.int32)
        pts = vertices.reshape((-1, 1, 2))
        cv2.polylines(image, [pts], isClosed=True, color=color, thickness=1)
        cv2.fillPoly(image, [pts], color=color)

            
image = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',triangle)

while True:
    cv2.imshow('image',image)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('b'):
        color = (255,0,0)
    elif k==ord('g'):
        color = (0,255,0)
    elif k==ord('r'):
        color = (0,0,255)
    elif k==ord('w'):
        color = (255,255,255)
    elif k==ord('y'):
        color = (51,255,255)
    elif k == 27:
        break
cv2.destroyAllWindows()

plt.imshow(image[...,::-1])
```
