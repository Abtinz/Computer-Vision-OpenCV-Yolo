```python
import cv2 as cv
from matplotlib import pyplot as plt
```

## Simple trditional Thresholding

```python
image = cv.imread('adad.jpg',0)

_ ,binary_threshold = cv.threshold(image, 127, 255, cv.THRESH_BINARY)

_ ,trunc = cv.threshold(image, 127, 255, cv.THRESH_TRUNC)

    
plt.figure(figsize=[10,10])
plt.subplot(131);plt.imshow(image, cmap='gray');plt.title("original image");
plt.subplot(132);plt.imshow(binary_threshold, cmap='gray');plt.title("Threshold - Binary");
plt.subplot(133);plt.imshow(trunc, cmap='gray');plt.title("trunc");
```

## Otsu

```python
image = cv.imread('adad.jpg',0)

_ ,binary_threshold = cv.threshold(image,127,255,cv.THRESH_BINARY)
_ ,otsu_image = cv.threshold(image,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
blured_image = cv.GaussianBlur(image,(5,5),0)
_ ,otsu_gaussian = cv.threshold(blured_image,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(binary_threshold, cmap='gray');plt.title("binary threshold")
plt.subplot(132);plt.imshow(otsu_image, cmap='gray');plt.title("otsu adaotive threshold")
plt.subplot(133);plt.imshow(otsu_gaussian, cmap='gray');plt.title("Gaussian noise deletion")
```

## Adaptive Mean Thresholding

```python
adaptive_c = cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,2)
plt.figure(figsize=[10,10])
plt.imshow(adaptive_c, cmap='gray');plt.title("Adaptive Mean Thresholding");
```

lets try noise free

```python
noise_free_image = cv.GaussianBlur(image,(15,15),0)
adaptive_c = cv.adaptiveThreshold(noise_free_image,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,7,2)
plt.figure(figsize=[10,10])
plt.imshow(adaptive_c, cmap='gray');plt.title("Adaptive Mean Thresholding on noise free image");
```

## Adaptive Gaussian Thresholding

```python
plt.figure(figsize=[10,10])
adaptive_gaussian = cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,2)
plt.imshow(adaptive_gaussian, cmap='gray');plt.title("Adaptive Gaussian Thresholding");
```

lets try noise free

```python
noise_free_image = cv.GaussianBlur(image,(15,15),0)
adaptive_gaussian = cv.adaptiveThreshold(noise_free_image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,2)
plt.figure(figsize=[10,10])
plt.imshow(adaptive_gaussian, cmap='gray');plt.title("Adaptive Gaussian on noise free image");
```
