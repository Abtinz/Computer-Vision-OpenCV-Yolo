```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

otsu will work based on the image histogram(the assumption of atsu)

in real world we have:

   * non-normal brigthness
    
   * shadows!

```python
image = cv.imread('sudoku.jpg',0)

_ ,binary_threshold = cv.threshold(image,127,255,cv.THRESH_BINARY)
_ ,otsu_image = cv.threshold(image,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
blured_image = cv.GaussianBlur(image,(3,3),0)
_ ,otsu_gaussian = cv.threshold(blured_image,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(binary_threshold, cmap='gray');plt.title("binary threshold")
plt.subplot(132);plt.imshow(otsu_image, cmap='gray');plt.title("otsu adaotive threshold")
plt.subplot(133);plt.imshow(otsu_gaussian, cmap='gray');plt.title("Gaussian noise deletion")
```
