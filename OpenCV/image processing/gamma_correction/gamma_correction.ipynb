# Imports and Basic Processes

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./Weeki_Wachee_spring_10079u.jpg')
plt.imshow(image[...,::-1])
```

# LUT

```python
look_up_table = np.empty((1,256) , "uint8")
look_up_table
```

here we will create  our LUT for gamma correction based on below formula:

    output = (i/255)^(gamma_coef) * 255

```python
from math import pow
gamma = 0.5
for i in range(256):
    look_up_table[0,i] = np.clip(pow(i/255.0 , gamma) * 255.0, 0, 255)
look_up_table
```

# Gamma and LUT Result

```python
#CV.LUT will map images colors and LUT for new image(we use it here for gamma correction)
out_put = cv.LUT(image , look_up_table)
plt.imshow(out_put[...,::-1])
```
