```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lets create our simple black board which we are going to draw others on
black_board = np.zeros((500,500,3), "uint8")

plt.imshow(black_board[...,::-1]); plt.title("black_board")
```

# Shapes

```python
#cv.line(image or board, starting cordinates of line, ending cordinates of line, color, thickness)
cv.line(black_board, (0,0), (500,50), (0,255,255), 5)
cv.line(black_board, (0,0), (50,500), (140,200,255), 5)
cv.line(black_board, (0,0), (500,500), (255,255,100), 5)
plt.imshow(black_board[...,::-1]);plt.title("lines")
```

Notes:
    
    1. color parameters are in BGR mode
    
    2. coordintes are x,y not height and width

```python
#cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

cv.rectangle(black_board, (150,150), (400,500), (0,0,255), 5)

plt.imshow(black_board[...,::-1]); plt.title("Rectangle");
```
