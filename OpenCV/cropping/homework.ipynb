# Imports

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

# Reading and Processing

```python
image = cv.imread("./sherlockholmes.jpg")
plt.title("sherlok holmes and doctor watson")
plt.xlabel("width")
plt.ylabel("heigth")
plt.imshow(image[:,:,::-1])
```

# sherlock head coordinates = (0,400) and (450,700)


```python
sherlock = image[0:400,450:700]
plt.title("sherlok holmes head")
plt.xlabel("width")
plt.ylabel("heigth")
plt.imshow(sherlock[:,:,::-1])
```

# watson head coordinates = (170,400) and (800,1000)

```python
watson = image[170:400,800:1000]
plt.title("doctor watson head")
plt.xlabel("width")
plt.ylabel("heigth")
plt.imshow(watson[:,:,::-1])
```

# Saving Section

```python
cv.imwrite("sherlock.jpg", sherlock)
cv.imwrite("watson.jpg", watson)
```
