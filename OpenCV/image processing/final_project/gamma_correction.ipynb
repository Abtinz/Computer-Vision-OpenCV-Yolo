# Imports and Basic Process

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./Weeki_Wachee_spring_10079u.jpg')
plt.imshow(image[...,::-1])
```

```python
from math import pow

"""
Gamma Correcction function:
    
    this funciton will create a LUT by given gamma coefficient and map given image with that. 
    args -> 
        image: taget image
        gamma: gamma coefficient
"""
def gamma_correction(image , gamma):
    
    look_up_table = np.empty((1,256) , "uint8")
    
    for i in range(256):
        look_up_table[0,i] = np.clip(pow(i/255.0 , gamma) * 255.0, 0, 255)
        
    out_put = cv.LUT(image , look_up_table)
    plt.imshow(out_put[...,::-1])
```

# Outputs

```python
gamma_correction(image , 0.25)
```

```python
gamma_correction(image , 0.33)
```

```python
gamma_correction(image , 0.5)
```

```python
gamma_correction(image , 1.5)
```

```python
gamma_correction(image , 2.0)
```
