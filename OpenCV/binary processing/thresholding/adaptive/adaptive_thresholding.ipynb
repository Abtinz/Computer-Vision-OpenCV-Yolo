```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
original_image = cv.imread('sudoku.jpg', 0)

free_noise_image = cv.medianBlur(original_image,5)
# we may use cv2.GaussianBlur(original_image,(5,5),0), the result is same!

plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(original_image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(free_noise_image, cmap='gray');plt.title("Noise removed image");
```

```python
#adaptive threshold over noisy image
_, binary_threshold = cv.threshold(original_image,127,255,cv.THRESH_BINARY)

#blocksize = 11 --> neighbors space dimension
#2 is a constant that is subtracted from the mean or weighted sum of the neighborhood pixels.
adaptive_c = cv.adaptiveThreshold(original_image,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,2)
adaptive_gaussian = cv.adaptiveThreshold(original_image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,2)


plt.figure(figsize=[15,10])

plt.subplot(131);plt.imshow(binary_threshold, cmap='gray');plt.title("Global Thresholding (v = 127)");
plt.subplot(132);plt.imshow(adaptive_c, cmap='gray');plt.title("Adaptive Mean Thresholding");
plt.subplot(133);plt.imshow(adaptive_gaussian, cmap='gray');plt.title("Adaptive Gaussian Thresholding");
```

```python

_,binary_threshold = cv.threshold(free_noise_image,127,255,cv.THRESH_BINARY)
adaptive_c = cv.adaptiveThreshold(free_noise_image,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,2)
adaptive_gaussian = cv.adaptiveThreshold(free_noise_image,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,2)

plt.figure(figsize=[15,10])


plt.subplot(131);plt.imshow(binary_threshold, cmap='gray');plt.title("Global Thresholding + noise free image");
plt.subplot(132);plt.imshow(adaptive_c, cmap='gray');plt.title("Adaptive Mean + noise free image");
plt.subplot(133);plt.imshow(adaptive_gaussian, cmap='gray');plt.title("Adaptive Gaussian + noise free image");
```
