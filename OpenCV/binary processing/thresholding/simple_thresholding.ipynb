```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('gradient.jpg')

binary_image = np.copy(image)
binary_image[binary_image <  127] = 0
binary_image[binary_image >= 127] = 255    

plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(binary_image, cmap='gray');plt.title("how legends calculate binary images");
```

## OpenCV Threshold Functions

Thresholding and inverted Thresholding

```python
_ ,binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
_ ,binary_inverted = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
plt.subplot(231);plt.imshow(binary, cmap='gray');plt.title("Threshold - Binary");
plt.subplot(232);plt.imshow(binary_inverted, cmap='gray');plt.title("Threshold - Binary Inverse")
```
