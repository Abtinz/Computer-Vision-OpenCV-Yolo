```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#second method: using imread at initial section for reading the image as grayscale image
image = cv.imread("./sherlockholmes.jpg")
plt.imshow(image[:,:,::-1])
```

let crop Holmes and Watson: width(400,1100)

```python
shrlock_john = image[:,400:1100]
cv.imwrite("shrlock_john.jpg" , shrlock_john)
plt.imshow(shrlock_john[:,:,::-1])
```

let crop the title: width(100,1380), heigth(700,1000) 

```python
title = image[700:1000,80:1360]
cv.imwrite("title.jpg" , title)
plt.imshow(title[:,:,::-1])
```
